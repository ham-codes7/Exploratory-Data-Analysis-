```python
import pandas as pd
import numpy as np
```

```python
python1 = pd.DataFrame({
    'StudentID':[101,103,105,107,109],
    'PythonScore':[85,76,90,65,88]
})

python2 = pd.DataFrame({
    'StudentID':[102,104,106,108,110],
    'PythonScore':[70,82,75,91,68]
})

datascience1 = pd.DataFrame({
    'StudentID':[101,102,103,105,107],
    'DSScore':[78,81,74,89,92]
})

datascience2 = pd.DataFrame({
    'StudentID':[104,106,108,109],
    'DSScore':[85,76,83,90]
})
```

```python
python = pd.concat([python1, python2], ignore_index=True)

datascience = pd.concat([datascience1, datascience2], ignore_index=True)

python
```

```python
datascience
```

```python
combined = pd.concat([python, datascience], axis=1)

combined
```

```python
inner = python.merge(datascience, how='inner')

inner
```

```python
left = python.merge(datascience, how='left')

left
```

```python
right = python.merge(datascience, how='right')

right
```

```python
sales = pd.DataFrame({
    'Company':['Apple','Samsung','Sony','Dell','HP'],
    'Product':['Phone','Tablet','TV','Laptop','Printer'],
    'Quantity':[5,8,3,6,4],
    'UnitPrice':[70000,25000,45000,80000,12000]
})

sales
```

```python
sales['TotalPrice'] = sales['Quantity'] * sales['UnitPrice']

sales
```

```python
sales['Company'].value_counts()
```

```python
sales['Company'].value_counts()
```

```python
data = np.arange(15).reshape((3,5))

indexers = ['Temperature','Humidity','Rainfall']

weather = pd.DataFrame(
    data,
    index=indexers,
    columns=['Delhi','Mumbai','Chennai','Bangalore','Hyderabad']
)

weather
```

```python
stacked = weather.stack()

stacked
```

```python
stacked.unstack()
```

```python
series1 = pd.Series(
    [100,200,300],
    index=['Math','Physics','Chemistry']
)

series2 = pd.Series(
    [400,500],
    index=['Biology','English']
)
```

```python
frame = pd.concat([series1, series2], keys=['Semester1','Semester2'])

frame
```

```python
frame.unstack()
```
