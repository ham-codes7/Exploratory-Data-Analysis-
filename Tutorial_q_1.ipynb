```python
import pandas as pd

# Load dataset
df = pd.read_csv("employee_salary.csv")

# Display first and last 5 records
print(df.head())
print(df.tail())

# Dataset shape
rows, columns = df.shape
print("Number of rows:", rows)
print("Number of columns:", columns)

# Metadata
print(df.columns)
print(df.dtypes)
df.info()

# Descriptive statistics
print(df.describe())

# Missing values
print(df.isnull().sum())

missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\nPercentage of Missing Values:")
print(missing_percentage)

print("\nColumn with Highest Missing Values:")
print(df.isnull().sum().idxmax())

print("Missing Value Count:")
print(df.isnull().sum().max())

# Duplicate records
duplicates = df.duplicated().sum()
print("Number of Duplicate Records:", duplicates)

# Dataset size before cleaning
print("Dataset Size Before Cleaning:", len(df))

# Remove duplicates
df = df.drop_duplicates()

# Dataset size after cleaning
print("Dataset Size After Cleaning:", len(df))
```
